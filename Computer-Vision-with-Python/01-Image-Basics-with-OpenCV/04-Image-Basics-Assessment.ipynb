#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/dog_backpack.png')
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
copy_img = img_rgb.copy()
flipped_img = cv2.flip(copy_img, 0)
plt.imshow(copy_img)
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
img_with_rect = img_rgb.copy()
cv2.rectangle(img_with_rect, pt1=(200, 380), pt2=(600, 700), color=(255, 0, 0), thickness=10)
plt.imshow(img_with_rect)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
vertices = np.array([[250, 700], [425, 400], [600, 700]], np.int32)
pts = vertices.reshape((-1, 1, 2))

img_with_triangle = img_rgb.copy()
cv2.polylines(img_with_triangle, [pts], isClosed=True, color=(0, 0, 255), thickness=20)
plt.imshow(img_with_triangle)
```

### BONUS TASK. Can you figure our how to fill in this triangle? It requires a different function that we didn't show in the lecture! See if you can use google search to find it.

[CLICK ME FOR A DIRECT LINK TO THE HINT](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
img_with_full_triangle = img_rgb.copy()
cv2.fillPoly(img_with_full_triangle, [pts], color=(0, 0, 255))
plt.imshow(img_with_full_triangle)
```

```python
import cv2
import numpy as np

WINDOW_NAME = 'Dog'


def draw_circle(event, x, y, flags,param):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x, y), 100, (0, 0, 255), 10)


def should_stop():
    return cv2.waitKey(20) & 0xFF == 27


img = cv2.imread('../DATA/dog_backpack.png')
cv2.namedWindow(winname=WINDOW_NAME)
cv2.setMouseCallback(WINDOW_NAME, draw_circle)
while not should_stop():
    cv2.imshow(WINDOW_NAME, img)
cv2.destroyAllWindows()
```
